### Fetching Dataset

```python
dataset_dir = "/content/drive/MyDrive/Dataset/Dataset_Thesis/"
# dataset = pd.read_csv(dataset_dir + "zilla64/zilla64.csv")
# dataset = pd.read_csv(dataset_dir + "BN-HW-DSN/BN-HW-DSN.csv")
dataset = pd.read_csv(dataset_dir + "BD_DB_64/BD_DB_64.csv")
```

```python
dataset
```

```python
X = dataset.iloc[:,0].values
y = dataset.iloc[:,1].values
print(X.shape)
print(y.shape)
```

### Preprocessing Image

```python
new_weight = 240
new_height = 80
for i, img_path in enumerate(X):
  # Reading Image
  # candidate_image = cv.imread(dataset_dir + "merged/raw/" + y[i] + "/" + img_path)
  candidate_image = cv.imread(dataset_dir + "BD_DB_64/labeled/" + y[i] + "/" + img_path)
  # Resizing Image
  candidate_rs = cv.resize(candidate_image, dsize=(new_weight, new_height))
  # Converting Image to Grayscale
  candidate_gray = cv.cvtColor(candidate_rs, cv.COLOR_BGR2GRAY)
  # Computing a Global Threshold Value from Grayscale Image(Using Otsu's Method) & Converting to Binary Image
  # ret, candidate_bw = cv.threshold(candidate_gray, 0, 255, cv.THRESH_BINARY + cv.THRESH_OTSU)
  ret, candidate_bw = cv.threshold(candidate_gray, 0, 255, cv.THRESH_BINARY_INV + cv.THRESH_OTSU)
  # Saving the processed Image
  # new_img_name = dataset_dir + "merged/converted_80_240/" + y[i] + "/" + img_path
  new_img_name = dataset_dir + "BD_DB_64/converted_80_240/" + y[i] + "/" + img_path
  cv.imwrite(new_img_name, candidate_bw)
  # print(img_path, end =" ")
```
